```python
import pandas as pd

# Load datasets
orders = pd.read_csv('../staging_2/orders/orders.csv')
order_items = pd.read_csv('../staging_2/order_items/order_items.csv')
products = pd.read_csv('../staging_2/products/products.csv')
categories= pd.read_csv('../staging_2/categories/categories.csv')
# Merge orders with order_items on order_id
merged_order_details = pd.merge(orders, order_items, on='order_id', how='inner')

# Merge with products on product_id
orders_products_details = pd.merge(merged_order_details, products, on='product_id', how='inner')
orders_products_details.drop(columns=['Extract Time', 'Source','extract_time_x','source_x','source_y','extract_time_y'], inplace=True)
orders_products_details = pd.merge(orders_products_details, categories, on='category_id', how='inner')
orders_products_details.drop(columns=['Extract Time', 'Source'], inplace=True)

# Save the merged dataset to a CSV file
orders_products_details.to_csv('../Information Mart/orders_products_details/orders_products_details.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Plot 1: Distribution of order status
plt.figure(figsize=(8, 6))
orders_products_details['order_status'].value_counts().plot(kind='bar')
plt.title('Distribution of Order Status')
plt.xlabel('Order Status')
plt.ylabel('Frequency')
plt.savefig('../Visualization/order_status_distribution.png')
plt.show()
```

```python
# Plot 2: Average quantity ordered per product category
plt.figure(figsize=(10, 6))
orders_products_details.groupby('category_name')['quantity'].mean().plot(kind='bar')
plt.title('Average Quantity Ordered per Product Category')
plt.xlabel('Product Category')
plt.ylabel('Average Quantity Ordered')
plt.xticks(rotation=45)
plt.savefig('../Visualization/avg_quantity_per_category.png')
plt.show()
```

```python
# Plot 3: Revenue generated by each product
plt.figure(figsize=(12, 6))
orders_products_details['revenue'] = orders_products_details['quantity'] * orders_products_details['list_price_egp']
revenue_by_product = orders_products_details.groupby('product_name')['revenue'].sum().sort_values(ascending=False).head(10)
revenue_by_product.plot(kind='bar')
plt.title('Top 10 Products by Revenue')
plt.xlabel('Product Name')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('../Visualization/top_products_by_revenue.png')
plt.show()
```
